```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Job_Market_India_with_Skills.csv")

# Clean column names
df.columns = df.columns.str.lower().str.strip()

df.head()
```

```python
# Total active job postings
total_jobs = len(df)

print("Total Job Demand (Active Postings):", total_jobs)
```

```python
top_companies = (
    df['company_name']
    .value_counts()
    .reset_index()
)

top_companies.columns = ['Company_name', 'Hiring_Count']

# Top 10 companies
top_10_companies = top_companies.head(10)

top_10_companies
```

```python
from collections import Counter

all_skills = []

for skills in df['skills_required']:
    if pd.notna(skills):
        skill_list = skills.lower().split(',')
        all_skills.extend([s.strip() for s in skill_list])

skill_demand = Counter(all_skills)

top_skills = pd.DataFrame(
    skill_demand.most_common(10),
    columns=['Skill', 'Demand_Count']
)

top_skills
```

```python
location_demand = (
    df['city']
    .value_counts()
    .reset_index()
)

location_demand.columns = ['Location', 'Job_Count']
                                                                                                                                                                                          
top_locations = location_demand.head(10)

top_locations
```

```python
experience_demand = (
    df['experience_level']
    .value_counts()
    .reset_index()
)

experience_demand.columns = ['Experience_Level', 'Job_Count']

experience_demand
```

```python
salary_by_exp = (
    df.groupby('experience_level')['salary_inr']
    .mean()
    .reset_index()
)

# Convert INR to LPA and round to nearest whole number
salary_by_exp['salary_lpa'] = (salary_by_exp['salary_inr'] / 100000).round(0).astype(int)

salary_by_exp
```
